```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
signal_1 = np.loadtxt('Data_1.txt')
signal_2 = np.loadtxt('Data_2.txt')
```

```python
peaks_max_1, _ = find_peaks(signal_1)
peaks_max_2, _ = find_peaks(signal_2)
```

```python
peaks_min_1, _ = find_peaks(-signal_1)
peaks_min_2, _ = find_peaks(-signal_2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(signal_1, label='Signal 1', color='black')
plt.plot(peaks_max_1, signal_1[peaks_max_1], 'ro', label='Maxima (Red)')
plt.plot(peaks_min_1, signal_1[peaks_min_1], 'bo', label='Minima (Blue)')
plt.title('Peaks in Signal 1')
plt.xlabel('Index')
plt.ylabel('Signal Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(signal_2, label='Signal 2', color='black')
plt.plot(peaks_max_2, signal_2[peaks_max_2], 'ro', label='Maxima (Red)')
plt.plot(peaks_min_2, signal_2[peaks_min_2], 'bo', label='Minima (Blue)')
plt.title('Peaks in Signal 2')
plt.xlabel('Index')
plt.ylabel('Signal Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Maxima indices for Signal 1:", peaks_max_1)
print("Minima indices for Signal 1:", peaks_min_1)
```

```python
print("Maxima indices for Signal 2:", peaks_max_2)
print("Minima indices for Signal 2:", peaks_min_2)
```
